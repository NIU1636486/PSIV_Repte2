```python
import cv2
import numpy as np

cap=cv2.VideoCapture('./output2.mp4')
# Coordinates for the region of interest (ROI)
pts = np.array([[359, 340],[186, 371] ,[8, 953], [446, 951]], np.int32)
pts = pts.reshape((-1, 1, 2))

last_frame = None

for i in range(100):
    ret,frame = cap.read()

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Create a mask for the region of interest (ROI)
    mask = np.zeros_like(frame, dtype=np.uint8)
    cv2.fillPoly(mask, [pts], (255, 255, 255))

    if last_frame is not None:
        # Compute the absolute difference only in the region of interest
        frame_difference = cv2.absdiff(frame, last_frame)*2
        
        
        # Apply the mask to show the difference only in the ROI
        frame_difference = cv2.bitwise_and(frame_difference, mask)
        print(frame_difference.mean())
        cv2.imshow('difference', frame_difference)

    # Display the current frame (can apply the mask here too if needed)
    if frame_difference.mean() > 1:
        cv2.putText(frame, 'YOLO ACTIVAT', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv2.LINE_AA)
    cv2.imshow('frame', frame)
    # draw pts

    break

    if cv2.waitKey(1) & 0xFF == 27:  # Exit on ESC key
        break


```
